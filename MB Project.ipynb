#### Dimensionality Reduction using PCA

```python
tr = trains.copy()
ts = tests.copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SC = StandardScaler()
```

```python
Scaledtrain = pd.DataFrame(SC.fit_transform(trains),columns=trains.columns)
```

```python
Scaledtrain
```

```python
from sklearn.decomposition import PCA
```

```python
tpca = PCA()
```

```python
Y = tpca.fit(Scaledtrain)
```

```python
var_cumu = np.cumsum(Y.explained_variance_ratio_)*100
```

```python
var_cumu
```

```python
k = np.argmax(var_cumu>95)
```

```python
print("No of components explaining  95% variance : ",k)
```

```python
plt.plot(var_cumu)
plt.show()
```

```python
Finaltrain = PCA(n_components = 149)
```

```python
train_dr = Finaltrain.fit_transform(Scaledtrain)
```

```python
train_dr
```

```python
data_train = pd.DataFrame(train_dr)
```

```python
data_train
```

```python
Scaledtest = pd.DataFrame(SC.fit_transform(tests),columns=tests.columns)
```

```python
Y = tpca.fit(Scaledtest)
```

```python
var_cumu = np.cumsum(Y.explained_variance_ratio_)*100
```

```python
k = np.argmax(var_cumu>95)
```

```python
print("No of components explaining  95% variance : ",k)
```

```python
plt.plot(var_cumu)
plt.show()
```

```python
Finaltest = PCA(n_components = 154)
```

```python
test_dr = Finaltrain.fit_transform(Scaledtest)
```

```python
data_test = pd.DataFrame(test_dr)
```

```python
data_test
```

Hence , the columns are reduced using dimensionality reduction using PCA with 95% variance

#### Predict Using XgBoost

```python
import xgboost as xgb
from sklearn.model_selection import cross_val_score,cross_val_predict
from xgboost import XGBRegressor
```

```python
data_train.head()
```

```python
x = data_train.drop(0,axis=1)
y= data_train[0]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.33,random_state=41)
```

```python
x_train.shape
```

```python
x_test.shape
```

### xgboost regressor is going to be used

```python
xg = XGBRegressor()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
xg.fit(x_train,y_train)
pred = xg.predict(x_test)
```

```python
print("r2_score(predicted_train) is ",r2_score(y_train,xg.predict(x_train)))
```

```python
print("Score of trained data",xg.score(x_train,y_train))
```

```python
print("Score of test data",xg.score(x_test,y_test))
```

```python
# mean absolute error
mean_absolute_error(y_test,pred)
```

```python
#mean_squared error
mean_squared_error(y_test,pred)
```

```python
# cross validation

score = cross_val_score(xg,x,y,cv=5)
print(xg,"cross val score is ",score)
print(score.mean())
```
